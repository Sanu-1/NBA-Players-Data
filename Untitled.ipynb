```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
f1 = pd.read_csv('mydata.csv')
```

```python
f1
```

```python
f1['Height']=np.random.randint(150,180,size=len(f1))
f1['Height']
```

```python
total_employee = f1['Team'].value_counts().sum()
```

```python
total_employee
```

```python
employee_distribution = f1['Team'].value_counts()
```

```python
employee_distribution
```

```python
percentege_split_of_employee = (employee_distribution)/(total_employee)*100
```

```python
percentege_split_of_employee
```

```python
employee_distribution.plot(kind='hist',color='green')
plt.xlabel('Number of players')
plt.ylabel('Teams')
plt.title('Players and their Team')
plt.show()
```

```python
employees = f1['Position'].value_counts()
employees
```

```python
employees.plot(kind='pie')
plt.legend(title = 'Players position', loc = 'lower right')
plt.show()
```

```python
predominant_age_group = f1['Age'].value_counts().idxmax()
predominant_age_group
```

```python
f1['Age'].plot(kind = 'hist',color = 'brown',edgecolor = 'black')
plt.xlabel('Age')
plt.ylabel('Number of players')
plt.title('Players age')
plt.show()
```

```python
highest_salary = f1.loc[f1['Salary'].idxmax()][['Name','Position','Team','Salary']]
highest_salary
```

```python
f1['Salary'].plot(kind='hist',color='green')
plt.xlabel('Salary in(thousands of dollars)')
plt.ylabel('Players')
plt.title('Salary of players')
plt.show()
```

```python
sns.scatterplot(x = f1['Age'],y = f1['Salary'],hue = f1['Salary'],data = 'mydata.csv')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Correlation of Salary and Age')
plt.show()
```

```python
correlation =  f1['Age'].corr(f1['Salary'])
correlation
```
